```python
# Drawing barplot
# Add more rows
# 3시간동안 80만 명이 들어오는 역 찾기

import csv
import matplotlib.pyplot as plt

f = open('C:\\Users\\User\Desktop\WORK\데이터분석\DATA\subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data); next(data)
mx = 0
mx_station = ''
result = []

for row in data:
    row[4:] = map(int, row[4:])
    result.append(sum(row[10:15:2]))
    if sum(row[10:15:2]) > mx :
        mx = sum(row[10:15:2])
        mx_station = row[3] + '(' + row[1] + ')'

result.sort()
plt.figure(dpi=300)
plt.bar(range(len(result)), result)
plt.show()
print(mx_station, mx)
```

```python
# 선택한 시간대 중 가장 많은 인원이 승차한 역과 명수 구하기
import csv

f = open('C:\\Users\\User\Desktop\WORK\데이터분석\DATA\subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data); next(data);

mx = 0
mx_station = ''
time = int(input("시간 >> "))

for row in data :
    row[4:] = map(int, row[4:])
    a = row[4 + (time-4) *2]
    if a > mx :
        mx = a
        mx_station = row[3] + '(' + row[1] + ')'

print(mx_station, mx)
```

```python
# 시간대별 최대 승차 인원 및 역 이름 구하기
import csv
import matplotlib.pyplot as plt

f = open('C:\\Users\\User\Desktop\WORK\데이터분석\DATA\subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data); next(data);

mx = [0] * 24
mx_station = [''] * 24

for row in data :
    row[4:] = map(int, row[4:])
    for j in range(24) :
        a = row[j*2+4]
        if a > mx[j] :
            mx[j] = a
            mx_station[j] = row[3] + '(' + str(j + 4) + ')'
    

print(mx_station)
print(mx)

plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), mx)
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
# 시간대별 최대 승차 인원 및 역 이름 구하기
import csv
import matplotlib.pyplot as plt

f = open('C:\\Users\\User\Desktop\WORK\데이터분석\DATA\subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data); next(data);

mx = [0] * 24
mx_station = [''] * 24

for row in data :
    row[4:] = map(int, row[4:])
    for j in range(24) :
        b = row[5 + j * 2]
        if b > mx[j] :
            mx[j] = b
            mx_station[j] = row[3] + '(' + str(j + 4) + ')'

plt.rc('font', family='Malgun Gothic')
plt.bar(range(24), mx, color='b')
plt.xticks(range(24), mx_station, rotation=90)
plt.show()
```

```python
# 시간대별 최대 승차 인원 및 역 이름 구하기
import csv
import matplotlib.pyplot as plt

f = open('C:\\Users\\User\Desktop\WORK\데이터분석\DATA\subwaytime.csv', encoding='cp949')
data = csv.reader(f)
next(data); next(data);

s_in = [0] * 24
s_out = [0] * 24

for row in data :
    row[4:] = map(int, row[4:])
    for i in range(24) :
        s_in[i] += row[4+i*2]
        s_out[i] += row[5+i*2]

plt.rc('font', family='Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label='승차')
plt.plot(s_out, label='하차')
plt.legend()
plt.xticks(range(24), range(4, 28))
plt.show()
```
