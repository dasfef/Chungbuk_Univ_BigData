```python
# 배열 초기화
import numpy as np

a = np.zeros(10)    # zero(0)로 이루어진 10개의 배열 개수
b = np.ones(10)     # one(1)로 이루어진 10개의 배열 개수
c = np.eye(10)      # n행 n열 배열 초기화

print(a, "\n")
print(b, "\n")
print(c, "\n")
```

```python
# np.arange 활용
import numpy as np

print(np.arange(3))         # 0 ~ n-1 까지
print(np.arange(3, 7))      # 3 ~ n-1 까지
print(np.arange(3, 7, 2))   # 3 ~ n-1 까지 2칸씩 띄어서
```

```python
# linspace 와 arange
import numpy as np

a = np.arange(1, 2, 0.1)        # 1 ~ 1.9 까지 0.1 간격
b = np.linspace(1, 2, 11)       # 1 ~ 2 까지 11개로 나눈 실수
print(a)
print(b)

_pi1 = np.arange(-np.pi, np.pi, np.pi/10)
_pi2 = np.linspace(-np.pi, np.pi, 20)
print(_pi1)
print(_pi2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

a = np.arange(-np.pi, np.pi, np.pi / 100)
plt.plot(a, np.sin(a))
plt.plot(a, np.cos(a))
plt.plot(a + np.pi / 2, np.sin(a))
plt.show()
```

```python
# mask 활용하기
import numpy as np

# mask 원리
a = np.arange(-5, 5)
print(a)
print(a < 0)
print(a[a<0])

# mask 응용
mask1 = abs(a) > 3
mask2 = abs(a) % 2 == 0
print(a[mask1])
print(a[mask2])

# mask 조건
print(a[mask1 + mask2])     # 둘 중 하나의 조건이라도 참일 경우
print(a[mask1 * mask2])     # 두 가지 조건이 모두 참일 경우
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(-100, 100, 1000)
y = np.random.randint(-100, 100, 1000)
mask1 = abs(x) > 50
mask2 = abs(y) > 50
x = x[mask1 + mask2]
y = y[mask1 + mask2]
size = np.random.rand(len(x + y)) * 100

plt.scatter(x, y, s=size, c=x, cmap='jet', alpha=0.3)
plt.colorbar()
plt.show()
```
