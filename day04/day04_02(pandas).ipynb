```python
import pandas as pd

df = pd.read_html('https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84', encoding='utf-8', header = 0, index_col = 0)

summer = df[0].iloc[:, :5]
summer.columns = ['하계참가횟수', '금', '은', '동', '계']
summer = summer.sort_values('금', ascending=False)
summer.to_excel('하계올림픽메달.xlsx')
# summer
```
