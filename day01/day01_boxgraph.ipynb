```python
import csv
f = open('seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)
result = []
aug = []
jan = []

for row in data :
    month = row[0].split('-')[1]
    if row[-1] != '' :
        if month == '08' :
            aug.append(float(row[-1]))
        if month == '01' :
            jan.append(float(row[-1]))

plt.boxplot([aug, jan])
# plt.boxplot(jan)
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

month = [[],[],[],[],[],[],[],[],[],[],[],[]]

for row in data :
    if row[-1] != '':
        idx = int(row[0].split('-')[1]) -1

        temp = float(row[-1])
        month[idx].append(temp)
    
print(len(month))

plt.boxplot(month)
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open('seoul.csv', encoding = 'cp949')
data = csv.reader(f)
next(data)

day = []
for i in range (31) :
    day.append([])

for row in data:
    if row[-1] != '':
        idx = int(row[0].split('-')[2]) -1
        temp = float(row[-1])
        if row[0].split('-')[1] == '08':
            day[idx].append(temp)

plt.style.use('ggplot')
plt.figure(figsize = (10, 5), dpi = 300)
plt.boxplot(day, showfliers = False)
plt.show()
```
