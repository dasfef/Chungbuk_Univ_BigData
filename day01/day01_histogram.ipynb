```python
import csv

f = open('C:\work\ML\day01\seoul.csv', 'r', encoding = 'cp949')
data = csv.reader(f, delimiter = ',')
header = next(data)
aug = []
jan = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
        if month == '01' :
            jan.append(float(row[-1]))

plt.hist(aug, bins = 100, color = 'r', label = 'Aug')
plt.hist(jan, bins = 100, color = 'b', label = 'Jan')
plt.legend()
plt.show()
```
